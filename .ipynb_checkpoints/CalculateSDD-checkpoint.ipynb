### Calculate Snow Disappearance Date from Time Series Temperature Data
Using thresholds for maximum daily temperature and daily temperature range, return snow disappearance dates where snow was present the previous day, then disappears. 

**Conditions for Snow Presence:**
1. Maximum daily temperature must be below be 1.5°C 
2. Diurnal temperature cycle (daily temperature range) must be below be 1.5°C 
3. Both of these conditions must be met for at least 48 hours.

Calculate daily maximum, minimum, and range for each sensor.

```python
CNFstats = pd.DataFrame()
CNFstats['1max'] = CNF['CNF1_temp'].groupby(CNF['Date']).max()
CNFstats['1min'] = CNF['CNF1_temp'].groupby(CNF['Date']).min()
CNFstats['1range'] = CNFstats['1max']-CNFstats['1min']

CNFstats['2max'] = CNF['CNF2_temp'].groupby(CNF['Date']).max()
CNFstats['2min'] = CNF['CNF2_temp'].groupby(CNF['Date']).min()
CNFstats['2range'] = CNFstats['2max']-CNFstats['2min']

CNFstats['3max'] = CNF['CNF3_temp'].groupby(CNF['Date']).max()
CNFstats['3min'] = CNF['CNF3_temp'].groupby(CNF['Date']).min()
CNFstats['3range'] = CNFstats['3max']-CNFstats['3min']

CNFstats['4max'] = CNF['CNF4_temp'].groupby(CNF['Date']).max()
CNFstats['4min'] = CNF['CNF4_temp'].groupby(CNF['Date']).min()
CNFstats['4range'] = CNFstats['4max']-CNFstats['4min']

CNFstats['5max'] = CNF['CNF5_temp'].groupby(CNF['Date']).max()
CNFstats['5min'] = CNF['CNF5_temp'].groupby(CNF['Date']).min()
CNFstats['5range'] = CNFstats['5max']-CNFstats['5min']

CNFstats['6max'] = CNF['CNF6_temp'].groupby(CNF['Date']).max()
CNFstats['6min'] = CNF['CNF6_temp'].groupby(CNF['Date']).min()
CNFstats['6range'] = CNFstats['6max']-CNFstats['6min']

CNFstats['7max'] = CNF['CNF7_temp'].groupby(CNF['Date']).max()
CNFstats['7min'] = CNF['CNF7_temp'].groupby(CNF['Date']).min()
CNFstats['7range'] = CNFstats['7max']-CNFstats['7min']

CNFstats['8max'] = CNF['CNF8_temp'].groupby(CNF['Date']).max()
CNFstats['8min'] = CNF['CNF8_temp'].groupby(CNF['Date']).min()
CNFstats['8range'] = CNFstats['8max']-CNFstats['8min']

CNFstats['9max'] = CNF['CNF9_temp'].groupby(CNF['Date']).max()
CNFstats['9min'] = CNF['CNF9_temp'].groupby(CNF['Date']).min()
CNFstats['9range'] = CNFstats['9max']-CNFstats['9min']
```

Set columns "SPx" to True only when both the daily max and daily range are below 1.5°C. 

```python
CNFstats['SP1'] = np.where((CNFstats['1max'] < 1.5) & (CNFstats['1range'] < 1.5),True, False)

CNFstats['SP2'] = np.where((CNFstats['2max'] < 1.5) & (CNFstats['2range'] < 1.5),True, False)

CNFstats['SP3'] = np.where((CNFstats['3max'] < 1.5) & (CNFstats['3range'] < 1.5),True, False)

CNFstats['SP4'] = np.where((CNFstats['4max'] < 1.5) & (CNFstats['4range'] < 1.5),True, False)

CNFstats['SP5'] = np.where((CNFstats['5max'] < 1.5) & (CNFstats['5range'] < 1.5),True, False)

CNFstats['SP6'] = np.where((CNFstats['6max'] < 1.5) & (CNFstats['6range'] < 1.5),True, False)

CNFstats['SP7'] = np.where((CNFstats['7max'] < 1.5) & (CNFstats['7range'] < 1.5),True, False)

CNFstats['SP8'] = np.where((CNFstats['8max'] < 1.5) & (CNFstats['8range'] < 1.5),True, False)

CNFstats['SP9'] = np.where((CNFstats['9max'] < 1.5) & (CNFstats['9range'] < 1.5),True, False)
```

To determine if the daily temperature max and range are below 1.5°C for 48 hours, compare the value of SPx for each row to that of  one row above and below. If the value of the current row is True and either the row before or after are also true, keep the value as True. Otherwise, set it to false. 

```python
CNFstats['SP1'] = np.where((CNFstats['SP1'] == True) & ((CNFstats['SP1'].shift(1) == True) | (CNFstats['SP1'].shift(-1) == True)), True, False)

CNFstats['SP2'] = np.where((CNFstats['SP2'] == True) & ((CNFstats['SP2'].shift(1) == True) | (CNFstats['SP2'].shift(-1) == True)), True, False)

CNFstats['SP3'] = np.where((CNFstats['SP3'] == True) & ((CNFstats['SP3'].shift(1) == True) | (CNFstats['SP3'].shift(-1) == True)), True, False)

CNFstats['SP4'] = np.where((CNFstats['SP4'] == True) & ((CNFstats['SP4'].shift(1) == True) | (CNFstats['SP4'].shift(-1) == True)), True, False)

CNFstats['SP5'] = np.where((CNFstats['SP5'] == True) & ((CNFstats['SP5'].shift(1) == True) | (CNFstats['SP5'].shift(-1) == True)), True, False)

CNFstats['SP6'] = np.where((CNFstats['SP6'] == True) & ((CNFstats['SP6'].shift(1) == True) | (CNFstats['SP6'].shift(-1) == True)), True, False)

CNFstats['SP7'] = np.where((CNFstats['SP7'] == True) & ((CNFstats['SP7'].shift(1) == True) | (CNFstats['SP7'].shift(-1) == True)), True, False)

CNFstats['SP8'] = np.where((CNFstats['SP8'] == True) & ((CNFstats['SP8'].shift(1) == True) | (CNFstats['SP8'].shift(-1) == True)), True, False)

CNFstats['SP9'] = np.where((CNFstats['SP9'] == True) & ((CNFstats['SP9'].shift(1) == True) | (CNFstats['SP9'].shift(-1) == True)), True, False)
```
